```python
from bs4 import BeautifulSoup
import requests
from itertools import groupby
import pandas as pd
```

```python
source = requests.get("https://en.wikipedia.org/wiki/List_of_London_Underground_stations").text
soup = BeautifulSoup(source, "lxml")

table = soup.find("table", class_="wikitable sortable plainrowheaders")
links = table.find_all("a")

table_data = []
for link in links:
    table_data.append(link.get("title"))

table_data = table_data[5:-2]
table_data

table_data = [data for data in table_data if data != None]
table_data

def generator_approach(list_):
    list_s=[]
    for i in list_:
        if 'Travelcard' in i:
            if list_s:
                yield list_s
            list_s=[]
        else:
            list_s.append(i)

    yield list_s

closure=generator_approach(table_data)
split_data = list(closure)
```

```python
station_names = pd.DataFrame(split_data)
station_names = station_names.rename(columns = {0: "Station"})["Station"]
```

```python
split_data

term = " line"
term2 = "line)"
term3 = "lines)"

tube_lines = [[ele for ele in sub if term in ele] for sub in split_data]
tube_lines = [[ele for ele in sub if term2 not in ele] for sub in tube_lines]
tube_lines = [[ele for ele in sub if term3 not in ele] for sub in tube_lines]

tube_lines = pd.DataFrame(tube_lines)
```

```python
result = pd.concat([station_names, tube_lines], axis=1)
result = result.fillna("")
result["Lines"] = result[0] + ";" + result[1] + ";" + result[2] + ";" + result[3] + ";" + result[4] + ";" + result[5]
result[["Station", "Lines"]]

# Dictionary Mapping
yes_no = {1: "Yes", 0: "No"}

# Creating Dummy Variables for Zones
result = result.join(result.pop("Lines").str.get_dummies(";").applymap(yes_no.get))

result.drop([0,1,2,3,4,5], axis=1, inplace=True)
result.head()
```
