# Predicting Footballers' Transfer Market Values

 # 1. Problem Framing

Within football getting the recruitment strategy correct is of the outmost importance on and off the pitch. Bringing the right player in at the correct price is often the difference between relegation, promotion, and winning competitions. While buying players at an overvaluation can lead to lack of resources in other areas.


The objective of this notebook will be to predict footballers transfer market values. This will culminate in a prediction model being constructed which will allow clubs to value players based on characteristics associated with them. Given the model within this notebook will be based upon players historic transfer market values this is an offline supervised learning problem. This means the models performance can be measured through comparing the models predictions to the labels. As there will be many features within this model and one prediction per player, this is both a multiple and univariate regression problem.

## 1.1. Measuring Model Performance

The performance measure which will be used to assess the quality of the models prediction in this notebook will be the Root Mean Square Error (RMSE). This common evaluation metric has been chosen as we would like to penalise predictions with high errors and be able to compare the units of the evaluation metric with the dependent variable.

## 1.2. Assumptions

There are 3 assumptions underpinning this notebook:
1.	The European Transfer Market differs from the other overseas markets. For example, the European and American Transfer Markets differ due to labour market restrictions, and transfer cap restrictions (Hoey, Peeters and Principe, 2020) enforced by the American soccer league. This has led us to only considering the Top 5 European Leagues (England, France, Germany, Italy, Spain).


2.	Defenders and Goalkeepers are valued differently from Strikers and Midfielders. Teams and fans often put more emphasis on players which score goals, as these win games. (Bleacher Report, 2013). Also, defensive statistics do not adequately reflect defensive players performances like goals and assists do for a strikers or midfielders. Given this is the case, this notebook will only investigate predicting Midfielders and Strikers transfer values.


3.	The Covid-19 pandemic has limited the effect of inflation on the European Football Transfer Market (CIES Football Observatory, 2020). Due to decreased funds available to clubs, transfer fees spent on players has not followed the yearly trends of rising. This had led to us considering players transfer market prices for the 2019 and 2020 seasons in one final dataset.

# 2. Getting the Data

The primary dataset within this notebook will be based on three publicly available data sources:

1.	Player Level Statistics based on the 2018/19 and 2019/20 seasons from FBref.com
2.	Club Level Statistics based on the 2018/19 and 2019/20 seasons from FBref.com
3.	Player Attribute information based on FIFA 20 and FIFA 21 datasets from Kaggle


These data sources will then be merged based upon mappings, and a Unique Identifier (UID). The first two data sources contain data that will be used as independent variables within the modelling process, while the third data source contains the dependent variable  footballers transfer value.


CIES Football Observatory (2020), explain the most important variables when taking a scientific approach to valuing a football player and this is illustrated in the figure below. Given its assumed that inflation has not had an affect on transfer prices 
the important areas to consider are: Player (first two sources) and Club factors (third source). 

The market values within the FIFA datasets are used as proxy for real world transfer market values. These prices should representative of real world prices, given that the FIFA series is associated with the world governing body of football.


```python
# Assigning location of Relevant Datasets to a variable (Team Mappings, Team Stats, Player Stats and Fifa Databases)
team_mappings_loc = 'Data/team_mappings.csv'
teamstats1819_loc = 'Data/big_5_teams_1819.csv'
teamstats1920_loc = 'Data/big_5_teams_1920.csv'
playerstats1819_loc = 'Data/big_5_players_1819.csv'
playerstats1920_loc = 'Data/big_5_players_1920.csv'
fifa20_loc = 'Data/players_20.csv'
fifa21_loc = 'Data/players_21.csv'

# Loading in Relevant Datasets as Pandas Dataframes
team_mappings_df = pd.read_csv(team_mappings_loc)
teamstats1819_df = pd.read_csv(teamstats1819_loc)
teamstats1920_df = pd.read_csv(teamstats1920_loc)
playerstats1819_df = pd.read_csv(playerstats1819_loc)
playerstats1920_df = pd.read_csv(playerstats1920_loc)
fifa20_df = pd.read_csv(fifa20_loc)
fifa21_df = pd.read_csv(fifa21_loc)
```

The player level statistics and club level statistics are combined to create two complete statistics datasets.


League Rank is chosen as this encapsulates a teams performance over a whole season while attendance (stadium capacity) is a good proxy for a clubs finances as the ticket sales make up a large percentage of club revenue.

```python
# Joining Relevant Team Stats (League Position and Stadium Attendance) to Player Stats
stats1819_df = pd.merge(playerstats1819_df, teamstats1819_df[["Squad", "LgRk", "Attendance"]], on="Squad")
stats1920_df = pd.merge(playerstats1920_df, teamstats1819_df[["Squad", "LgRk", "Attendance"]], on="Squad")
```

The club names within the statistics datasets are mapped to the FIFA format of club names in a mapping file.

```python
stats_dfs = [stats1819_df, stats1920_df]

# Looping over the stats datasets
for df in stats_dfs:

    # Aligning Stats Datasets' Team Naming Convention with Fifa datasets
    df["Squad"] = df["Squad"].map(team_mappings_df.set_index("Team Stats Name")["Fifa Name"])

    # Splitting Player Column to extract the last name
    df["Player"] = df.Player.str.split("\\", expand=True)[1]
    df["Player"] = df.Player.str.split("-").str[-1]
    
    # Creating a Unique Identifier Column with Player Name, Team and Birth Year
    df["UID"] = df["Player"] + "-" + df["Squad"] + "-" + df["Born"].astype(str)

# Removing Irrelevant Columns (Columns unrelated to performance statistics and UID) 
stats1819_df = stats1819_df.iloc[: , np.r_[3, 5, 8:36]]
stats1920_df = stats1920_df.iloc[: , np.r_[3, 5, 8:36]]
```

The only columns selected from the FIFA datasets are player age, player transfer market value and UID. The UID will be used to join the statistics datasets with the FIFA datasets.

```python
fifa_dfs = [fifa20_df, fifa21_df]
fifa_cols = ["UID", "age", "value_eur"]

# Looping over the Fifa Datasets
for df in fifa_dfs:
    
    # Splitting Short Name Column to extract the last name
    names = df.short_name.str.split(".", expand=True)
    names = df.short_name.str.split(" ", expand=True)
    df["short_name"] = names.ffill(axis=1).iloc[:, -1]

    # Splitting Date of Birth Column to extract Birth Year
    df["Born"] = df.dob.str.split("-", expand=True)[0]

    # Creating a Unique Identifier Column with Player Name, Team and Birth Year
    df["UID"] = df["short_name"] + "-" + df["club"] + "-" + df["Born"].astype(str)
    
# Removing Irrelavnt Columns (Columns unrelated to Age, Market Value and UID)
fifa20_df = fifa20_df[fifa_cols]
fifa21_df = fifa20_df[fifa_cols]
```

FIFA data is merged with statistics data from the previous season, as transfer values are typically based on the prior season. However, it must be noted that rows of data were lost in merging process, as some player names were formatted differently across the two datasets. 

Also, Goalkeepers and Defenders are not considered within our dataset as per our 2nd assumption.

```python
# Joining and concatenating the statistics and Fifa Datasets
season20_df = pd.merge(fifa20_df, stats1819_df, on="UID")
season21_df = pd.merge(fifa21_df, stats1920_df, on="UID")
season2021_df = pd.concat([season20_df, season21_df])

# Aggregating Positions Variable
# Creating Position Flags (Columns) specifcying a players primary position
season2021_df["GK"] = list(map(lambda x: x.startswith('GK'), season2021_df['Pos']))
season2021_df["DF"] = list(map(lambda x: x.startswith('DF'), season2021_df['Pos']))
season2021_df["MF"] = list(map(lambda x: x.startswith('MF'), season2021_df['Pos']))
season2021_df["FW"] = list(map(lambda x: x.startswith('FW'), season2021_df['Pos']))

# Creating a condition which replaces the players current positions with only their primary position
conditions = [
    (season2021_df["GK"] == True),
    (season2021_df["DF"] == True),
    (season2021_df["MF"] == True),
    (season2021_df["FW"] == True)
]

values = ["GoalKeeper", "Defender", "Midfielder", "Attacker"]

season2021_df["Pos"] = np.select(conditions, values)

# Removing Position Flags (Columns)
season2021_df = season2021_df.iloc[: , :-4]

# Filtering out Goalkeepers and DefendersFIFA data is merged with statistics data from the previous season, as transfer values are typically based on the prior season. For example, Player Xs value at the beginning of 2021 should be related to their performances in 2019/20 season. However, it must be noted that rows of data were lost in merging process, as some player names were formatted differently across the two datasets. Also, players with the positions of Goalkeeper and Defenders are not considered within our dataset as per our 2nd assumption.
season2021_df = season2021_df[(season2021_df["Pos"] == "Attacker") | (season2021_df["Pos"] == "Midfielder")]
```

# 4. Preparing the Data for Machine Learning Models

This section will put the cleaned dataset through pipelines in order to have dataset ready for machine learning algorithms. These pipelines include feature engineering and standard scaling of numerical attributes and one-hot encoding of categorical attributes.

```python
# Splitting train set into predictors and labels
x_train = train_set.drop("value_eur", axis=1)
y_train = train_set["value_eur"].copy()
y_train = y_train.to_numpy()
```

```python
# Checking nulls in train predictors and labels set
print(any(x_train.isnull().any()))
print(np.isnan(np.sum(y_train)))
```

```python
# Setting categorical attributes columns
cat1 = x_train.columns.get_loc("Pos")
cat2 = x_train.columns.get_loc("Comp")
cat_attribs = [cat1, cat2]

# Setting empty list for numercial attributes columns
num_attribs = []

# Retriving number of columns in train set
cols = len(x_train.columns)

# Looping through columns and adding to numerical attributes columns list
for x in range(0, cols):
    num_attribs.append(x)

# Removing categorical attributes from numerical attributes list
num_attribs = [x for x in num_attribs if x not in cat_attribs] 
```

```python
# Selecting correct columns indices for additional attributes (ouput by age & club size)
age_ix, output_ix, attendance_ix, LgRk_ix = [list(x_train.drop(["Pos", 'Comp'], axis=1).columns).index(col)
                                             for col in ("age", "Expected npxG+xA", "Attendance", "LgRk")]

# [FEATURE ENGINEERING) Defining function which creates addtional attributes (output by age & club size)
def add_new_features(data):
    output_by_age = (data[:, output_ix])**2 / data[:, age_ix]
    club_size = (data[:, attendance_ix])**2 / data[:, LgRk_ix]
    return np.c_[data, output_by_age, club_size]
```

The new features for output by age and club size are added to the numerical pipeline. These features should have strong predicting power as younger players who perform well are always in higher demand, and bigger clubs tend to have better players who are more expensive.

```python
# Setting pipeline for numerical variables
num_pipeline = Pipeline(steps = [
    ('attribs_adder', FunctionTransformer(add_new_features, validate=False)),
    ('std_scaler', StandardScaler())
    ])
```

```python
# Applying relevant pipelines to specified attributes
full_pipeline = ColumnTransformer([
    ("numerical", FunctionTransformer(add_new_features, validate=False), num_attribs),
    ("categorical", OneHotEncoder(), cat_attribs), # Creating dummy variables based off categorical variables
        ])

# Transforming training predictors
x_train_prepped = full_pipeline.fit_transform(x_train.values)
x_train_prepped
```

The cleaned dataset is transformed into a dataset ready for machine learning algorithms, with each feature and its transformation listed below.

```python
# Setting up stratified cross validation
skf = StratifiedKFold(n_splits=3, shuffle=True, random_state=96) 

# Empty lists to store results of stratified cross validations
lin_reg_accu = [] 
lasso_reg_accu = []
ridge_reg_accu = []
el_reg_accu = []
dt_reg_accu = []
rf_reg_accu = []
gb_reg_accu = []
xgb_reg_accu = []

# Lists of regression functions and empty stratified cross validations
regs = [lin_reg, lasso_reg, ridge_reg, el_reg, dt_reg, rf_reg, gb_reg, xgb_reg]
regs_accu = [lin_reg_accu, lasso_reg_accu, ridge_reg_accu, el_reg_accu, dt_reg_accu, rf_reg_accu, gb_reg_accu, xgb_reg_accu]

# Looping over regression functions to fill empty lists with cross validation scores
for reg, reg_accu in zip(regs, regs_accu):

    for train_index, test_index in skf.split(x_train_prepped, y_train):
    
        x_train_fold, x_test_fold = x_train_prepped[train_index], x_train_prepped[test_index] 
        y_train_fold, y_test_fold = y_train[train_index], y_train[test_index] 
        reg.fit(x_train_fold, y_train_fold) 
        reg_accu.append(reg.score(x_test_fold, y_test_fold))
```

This attempt at stratified cross validation across the models will not be effective, as there are too few of one particular class in the independent variable. This means a cross validation approach will be taken below.

```python
# Empty lists to store scores of cross validations
lin_reg_accu = [] 
lasso_reg_accu = []
ridge_reg_accu = []
el_reg_accu = []
dt_reg_accu = []
rf_reg_accu = []
gb_reg_accu = []
xgb_reg_accu = []

# Lists of regression functions and empty cross validations lists
regs = [lin_reg, lasso_reg, ridge_reg, el_reg, dt_reg, rf_reg, gb_reg, xgb_reg]
regs_accu = [lin_reg_accu, lasso_reg_accu, ridge_reg_accu, el_reg_accu, dt_reg_accu, rf_reg_accu, gb_reg_accu, xgb_reg_accu]

# Looping over regression functions to fill empty lists with cross validation scores
for reg, reg_accu in zip(regs, regs_accu):
    scores = cross_val_score(reg, x_train_prepped, y_train, scoring="neg_mean_squared_error", cv=3)
    scores = np.sqrt(-scores)
    reg_accu.append(scores)
```

```python
# Empty List for feature importance (coefficients)
feature_imp = []

# List of relevant regression functions
regs = [lin_reg, lasso_reg, ridge_reg, el_reg, dt_reg, rf_reg, gb_reg, xgb_reg]

# Adding relevant models feature importance to the empty feature importance list
for reg in regs:
    try:
        feature_imp.append(reg.coef_)
    except:
        feature_imp.append(reg.feature_importances_)
    
# Empty sub-lists for feature importance (coefficients) for each relevant model
lin_reg_coef = []
lasso_reg_coef = []
ridge_reg_coef = []
el_reg_coef = []
dt_reg_coef = []
rf_reg_coef = []
gb_reg_coef = []
xgb_reg_coef = []

# List of sub-lists for each relevant model's feature importance (coefficients)
sub_feature_imp = [lin_reg_coef, lasso_reg_coef, ridge_reg_coef, el_reg_coef, dt_reg_coef, rf_reg_coef, gb_reg_coef, xgb_reg_coef]    

# Appending feature importance (coefficient) values to their relevant sub-list
for sub, x in zip(sub_feature_imp, range(len(sub_feature_imp))):
                  sub.append(feature_imp[x])
```

```python
# Setting format of DataFrame to round to 2 decimal places
pd.options.display.float_format = '{:.2f}'.format

# Creating dataframe for Cross Validation Scores
regs_accu_df = pd.DataFrame(data = np.concatenate(regs_accu))

# Renaming index and columns of Dataframe
regs_accu_df = regs_accu_df.rename(index={0: "Linear", 1: "Lasso", 2: "Ridge", 3: "Elastic Net", 4: "Decision Tree", 5: "Random Forest", 6: "Gradient Boosting", 7: "XGBoost"})
regs_accu_df = regs_accu_df.rename(columns = {0: "Fold 1", 1: "Fold 2", 2: "Fold 3"})

# Adding columns for the Mean and Standard Deviation
regs_accu_df['Mean'] = regs_accu_df.mean(axis = 1)
regs_accu_df['Stand Dev'] = regs_accu_df[["Fold 1", "Fold 2", "Fold 3"]].std(axis = 1)

# Adding columns for Mean and Standard Deviation Rank
regs_accu_df['Mean Rank'] = regs_accu_df['Mean'].rank(ascending = True)
regs_accu_df['Stand Dev Rank'] = regs_accu_df['Stand Dev'].rank(ascending = True) 

# DataFrame of Cross Validation Scores
regs_accu_df.sort_values(by = ['Mean Rank', 'Stand Dev Rank'])
```

The Random Forest Model is the best model with the lowest RMSE, however the top three models are in the bottom four for RMSE standard deviation. This means they are not consistently getting their lower RMSE scores.

# 6. Fine Tuning Models

Here we will aim to improve the performance of models by fine tuning different combinations of hyperparameters  with cross-validation in order to reduce overfitting. First, a random grid search will be used to narrow the down selection space of hyperparameters, followed by a grid search. Upon finding the best combination of hyperparameters, these models will be combined in a voting ensemble regressor.

## 6.1. Gradient Boosting Regression

```python
# Fitting the Gradient Boosting Regression Model
gb_reg = GradientBoostingRegressor(random_state=96)

# Viewing Gradient Boosting Parameters
print('Gradient Boosting Parameters in use:\n')
pprint(gb_reg.get_params())
```

```python
# Setting the loss function to be optimised
loss = ["ls"]

# Seting the learning rate to be used
learning_rate = [0.001, 0.01, 0.1, 0.2, 0.3]

# Setting the number of boosting stages
n_estimators = [int(i) for i in np.linspace(start = 100, stop = 500, num = 5)]

# Setting the number of features to consider when searching for the best split
max_features = ['auto', 'sqrt']

# Setting the maximum number of levels of regression estimators
max_depth = [5, 10, 15, 20]

# Setting the minimum number of samples required to split an internal node
min_samples_split = [2, 5, 10]

# Setting the minimum number of samples required to be a leaf node
min_samples_leaf = [1, 2, 4]

# Inputting range for each parameter into the random grid search
rand_grid = {'loss': loss,
             'learning_rate': learning_rate,
             'n_estimators': n_estimators,
             'max_features': max_features,
             'max_depth': max_depth,
             'min_samples_split': min_samples_split,
             'min_samples_leaf': min_samples_leaf}

# Viewing random grid search parameters
pprint(rand_grid)
```

```python
# Randomly searching parameters, with 3-fold cross validation, across 900 combinations
rand_gb = RandomizedSearchCV(estimator = gb_reg, param_distributions = rand_grid, n_iter = 900, 
                             cv = 3, verbose=2, random_state = 96 , n_jobs = -1)

# Fitting the random search models
rand_gb.fit(x_train_prepped2, y_train)

# Vieweing the Best Parameters
pprint(rand_gb.best_params_)
```

```python
# Inputting range for each parameter based on best parameters of random search
param_grid = {
    "learning_rate": [0.01, 0.02, 0.05],
    "loss" : ["ls"],
    "max_depth": [18, 19, 20],
    "max_features": ["sqrt"],
    "min_samples_leaf": [1, 2, 3],
    "min_samples_split": [2, 3, 4],
    "n_estimators": [450, 500, 550]
}

# Starting the grid search models
grid_gb = GridSearchCV(estimator = gb_reg, param_grid = param_grid,
                       cv = 3, n_jobs = -1, verbose = 2)

# Fitting the grid search models
grid_gb.fit(x_train_prepped2, y_train)

# Vieweing the Best Parameters
pprint(grid_gb.best_params_)
```

```python
# Fitting the Base Gradient Boosting Regression Model
gb_reg_base = RandomForestRegressor(random_state=96)
gb_reg_base.fit(x_train_prepped2, y_train) 

# Fitting the Best Gradient Boosting Regression Model
gb_reg_best = grid_gb.best_estimator_
gb_reg_best.fit(x_train_prepped2, y_train) 

# Empty lists to store scores of cross validations
gb_base_accu = []
gb_best_accu = []

# Lists of Gradient Boosting functions and empty cross validations lists
gbs = [gb_reg_base, gb_reg_best]
gbs_accu = [gb_base_accu, gb_best_accu]

# Looping over Gradient Boosting functions to fill empty lists with cross validation scores
for gb, gb_accu in zip(gbs, gbs_accu):
    scores = cross_val_score(gb, x_train_prepped2, y_train, scoring="neg_mean_squared_error", cv=3)
    scores = np.sqrt(-scores)
    gb_accu.append(scores)

# Creating dataframe for Cross Validation Scores
gb_accu_df = pd.DataFrame(data = np.concatenate(gbs_accu))

# Renaming index and columns of Dataframe
gb_accu_df = gb_accu_df.rename(index={0: "Base Model", 1: "Best Model"})
gb_accu_df = gb_accu_df.rename(columns = {0: "Fold 1", 1: "Fold 2", 2: "Fold 3"})

# Adding columns for the Mean and Standard Deviation
gb_accu_df['Mean'] = gb_accu_df.mean(axis = 1)
gb_accu_df['Stand Dev'] = gb_accu_df[["Fold 1", "Fold 2", "Fold 3"]].std(axis = 1)

# Adding columns for Mean and Standard Deviation Rank
gb_accu_df['Mean Rank'] = gb_accu_df['Mean'].rank(ascending = True)
gb_accu_df['Stand Dev Rank'] = gb_accu_df['Stand Dev'].rank(ascending = True) 

# DataFrame of Cross Validation Scores
gb_accu_df = gb_accu_df.sort_values(by = ['Mean Rank', 'Stand Dev Rank']) 
gb_accu_df
```

```python
# Calculating percentage differences in RMSE Mean and Standard Deviation
gb_accu_df = gb_accu_df[["Mean", "Stand Dev"]]
gb_mean_diff, gb_stand_dev_diff = gb_accu_df.loc["Best Model"]/gb_accu_df.loc["Base Model"]

# Calculating difference from the Best Model to Base Model
gb_mean_diff = round((1 - gb_mean_diff)*100, 2)
gb_stand_dev_diff = round((1 - gb_stand_dev_diff)*100, 2)

print("The Gradient Boosting Best Model's Mean RMSE is", str(gb_mean_diff), "% lower than the Base Model's Mean RMSE")
print("The Gradient Boosting Best Model's RMSE Standard Deviation is", str(gb_stand_dev_diff), "% lower than the Base Model's RMSE Standard Deviation")
```

```python
# Fitting the Random Forest Regression Model
rf_reg = RandomForestRegressor(random_state=96)

# Viewing Random Forest Parameters
print('Random Forest Parameters in use:\n')
pprint(rf_reg.get_params())
```

```python
# Setting the number of trees in the forest
n_estimators = [int(i) for i in np.linspace(start = 100, stop = 500, num = 5)]

# Setting the number of features to consider when searching for the best split
max_features = ['auto', 'sqrt']

# Setting the maximum number of levels the tree
max_depth = [5, 10, 15, 20]

# Setting the minimum number of samples required to split an internal node
min_samples_split = [2, 5, 10]

# Setting the minimum number of samples required to be a leaf node
min_samples_leaf = [1, 2, 4]

# Setting whether bootstraps smaples are used in tree building
bootstrap = [True, False]

# Inputting range for each parameter into the random grid search
rand_grid = {'n_estimators': n_estimators,
             'max_features': max_features,
             'max_depth': max_depth,
             'min_samples_split': min_samples_split,
             'min_samples_leaf': min_samples_leaf,
             'bootstrap': bootstrap}

# Viewing random grid search parameters
pprint(rand_grid)
```

## 6.4. Voting Ensemble Regression

This voting regressor will take the best models from Gradient Boosting, Random Forests, and XGBoost. These models will then be combined at different weights and the best model in terms of Mean RMSE and standard deviation will be selected as the final model.

```python
# Setting Weight Combinations between the 3 best models
weights = [[1, 1, 0], [1, 0, 1], [0, 1, 1],
           [2, 1, 1], [1, 2, 1], [1, 1, 2], 
           [2, 2, 1], [2, 1, 2], [1, 2, 2], None]

# Empty lists to store scores of cross validations
vote_1 = []
vote_2 = []
vote_3 = []
vote_4 = []
vote_5 = []
vote_6 = []
vote_7 = []
vote_8 = []
vote_9 = []
vote_10 = []

# Lists empty cross validations score lists
voting_scores = [vote_1, vote_2, vote_3, vote_4, vote_5, vote_6, vote_7, vote_8, vote_9, vote_10]

# Looping over weight combinations to fill empty lists with cross validation scores
for w, v in zip(weights, voting_scores):
    vote_reg = VotingRegressor(estimators=[('gb', gb_reg_best),
                                           ('fr', rf_reg_best),
                                           ('xgb', xgb_reg_best)],
                               weights = w)
    scores = cross_val_score(vote_reg, x_train_prepped2, y_train, scoring="neg_mean_squared_error", cv=3)
    scores = np.sqrt(-scores)
    v.append(scores)

# Creating dataframe for Cross Validation Scores
voting_scores_df = pd.DataFrame(data = np.concatenate(voting_scores))
voting_scores_df

# Renaming index and columns of Dataframe
voting_scores_df = voting_scores_df.rename(index={0: "Vote 1", 1: "Vote 2", 2: "Vote 3", 3: "Vote 4", 4: "Vote 5", 5: "Vote 6", 6: "Vote 7", 7: "Vote 8", 8: "Vote 9", 9: "Vote 10"})
voting_scores_df = voting_scores_df.rename(columns = {0: "Fold 1", 1: "Fold 2", 2: "Fold 3"})

# Adding columns for the Mean and Standard Deviation
voting_scores_df['Mean'] = voting_scores_df.mean(axis = 1)
voting_scores_df['Stand Dev'] = voting_scores_df[["Fold 1", "Fold 2", "Fold 3"]].std(axis = 1)

# Adding columns for Mean and Standard Deviation Rank
voting_scores_df['Mean Rank'] = voting_scores_df['Mean'].rank(ascending = True)
voting_scores_df['Stand Dev Rank'] = voting_scores_df['Stand Dev'].rank(ascending = True) 

# DataFrame of Cross Validation Scores
voting_scores_df = voting_scores_df.sort_values(by = ['Mean Rank', 'Stand Dev Rank']) 
voting_scores_df
```

Our final model is Vote 2 as it has the lowest Mean RMSE. Its RMSE standard deviation rank of 8 is acceptable when its compared againsnt the other top 4 models.

```python
# Concatenating Vote 8 Ensemble Regression with Best Performing Individual Models
comparison = [voting_scores_df, gb_accu_df, rf_accu_df, xgb_accu_df]
comparison_df = pd.concat(comparison)

# Selecting relevant columns and rows from combined dataframe
comparison_df = comparison_df.iloc[[0, 10, 12, 14]]
comparison_df = comparison_df[["Mean", "Stand Dev"]]

# Renaming indexes of dataframe
comparison_df.reset_index(inplace = True)
comparison_df.drop("index", inplace = True, axis = 1)
comparison_df = comparison_df.rename(index={0: "Vote 2", 1: "Gradient Boost", 2: "Random Forest", 3: "XGBoost"})

# Adding columns for Mean and Standard Deviation Rank
comparison_df['Mean Rank'] = comparison_df['Mean'].rank(ascending = True)
comparison_df['Stand Dev Rank'] = comparison_df['Stand Dev'].rank(ascending = True) 

# DataFrame of Cross Validation Scores
comparison_df = comparison_df.sort_values(by = ['Mean Rank', 'Stand Dev Rank']) 
comparison_df
```

The final model's results are not strong, given the mean RMSE is fairly close to the average player's transfer market value. This means that players below the average transfer market value (approximately 12 million euros) may be valued at approximately double their worth in this final model. In a situation where there is more time and resources a business would benefit from manually checking the UID's (player names) of each player as this would lead to fewer observations being lost in the merging process of data sources.

Further feature engineering may prove beneficial, as the new features of club size and player output by age proved to be two of the most influential variables within the final model.
